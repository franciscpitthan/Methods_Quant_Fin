```python
import pandas as pd
import sqlite3

df_path='C:/Users/u0135988/OneDrive - KU Leuven/Research/Quant/Methods_Quant_Fin/Example - Fraud Detection - Kaggle/fraudTest.csv'
db_path='C:/Users/u0135988/OneDrive - KU Leuven/Research/Quant/Methods_Quant_Fin/Example - Fraud Detection - Kaggle/fraud.db'


#Read the csv file called df for simplicity
df = pd.read_csv(df_path)

#Create a SQLite database
conn = sqlite3.connect(db_path) 
cursor = conn.cursor()

#Upload data to the SQLite database
table_name = 'fraud_data'
#df.to_sql is useful to write an existing data frame (here df) to sql
df.to_sql(table_name, conn, if_exists='replace', index=True) #index is to write df's idex as column

#Run queries on the database
query = f"SELECT * FROM {table_name} LIMIT 20;"  
#This pd function allows to execute a query and save outcome 
result = pd.read_sql_query(query, conn)
print(result)

# Close the connection
conn.close()
```
